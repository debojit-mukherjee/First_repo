```python
# Stage 3: Implementation with Exercises + Feedback

def greet_user(name):
    return f"Hello {name}, I am your AI Chatbot! I can teach SQL, Python, Java, C++, and HTML."

def choose_topic():
    topic = input("Which topic would you like to learn? (SQL/Python/Java/C++/HTML): ")
    return topic

def explain_topic(topic):
    if topic.lower() == "python":
        return "Python is beginner-friendly. Let's start with variables!"
    elif topic.lower() == "sql":
        return "SQL is used to manage databases. Let's start with SELECT queries!"
    elif topic.lower() == "java":
        return "Java is powerful for applications. Let's start with classes!"
    elif topic.lower() == "c++":
        return "C++ is great for performance-heavy apps. Let's start with functions!"
    elif topic.lower() == "html":
        return "HTML is the backbone of web pages. Let's start with headings!"
    elif topic.lower() == "api":
        return "APIs let applications talk to each other. Let's start with GET requests!"
    else:
        return "Sorry, I don’t know that topic yet."

def practice_exercise(topic):
    if topic.lower() == "python":
        import random
        python_exercises = [
            ("What will this print?\nprint(2*3)", "6"),
            ("What keyword defines a function?", "def"),
            ("What symbol is used for comments?", "#")
        ]

        question, correct_answer = random.choice(python_exercises)
        answer = input(f"Exercise: {question}\nYour answer: ")
        if answer.strip().lower() == correct_answer.lower():
          print("Correct! 🎉")
        else:
          print(f"Try again! The answer is {correct_answer}.")


    elif topic.lower() == "sql":
        answer = input("Exercise: Which SQL command retrieves data?\nYour answer: ")
        if answer.lower().strip() == "select":
            print("Correct! 🎉")
        else:
            print("Try again! The answer is SELECT.")

    elif topic.lower() == "java":
        answer = input("Exercise: What keyword is used to define a class in Java?\nYour answer: ")
        if answer.lower().strip() == "class":
            print("Correct! 🎉")
        else:
            print("Try again! The answer is 'class'.")

    elif topic.lower() == "c++":
        answer = input("Exercise: Which symbol is used for single-line comments in C++?\nYour answer: ")
        if answer.strip() == "//":
            print("Correct! 🎉")
        else:
            print("Try again! The answer is //.")

    elif topic.lower() == "html":
        answer = input("Exercise: Which tag is used for the largest heading in HTML?\nYour answer: ")
        if answer.lower().strip() == "<h1>":
            print("Correct! 🎉")
        else:
            print("Try again! The answer is <h1>.")

    else:
        print("Exercises for this topic are coming soon!")

# Stage 4 + 5: Run interactively in Colab
def chatbot():
    name = input("Enter your name: ")
    print(greet_user(name))

    topic = choose_topic()
    print(explain_topic(topic))

    practice_exercise(topic)

# Start the chatbot
chatbot()
```
